# Headroom: Before / After

This notebook demonstrates Headroom in the clearest way possible for a video:
- first the raw context
- then the same context after Headroom
- same output format
- visible token savings


## 1) Three integration modes

- `compress()` = call it directly in your code
- `proxy` = Headroom runs as a local intermediary server
- `SDK / wrapper` = Headroom wraps your existing client

This notebook uses only `compress()` because it makes the before/after contrast easiest to see.


```python
import json

import tiktoken
from headroom import compress

MODEL = "gpt-4o"

def token_count(messages, model=MODEL):
    try:
        enc = tiktoken.encoding_for_model(model)
    except Exception:
        enc = tiktoken.get_encoding("o200k_base")
    payload = json.dumps(messages, ensure_ascii=False, sort_keys=True)
    return len(enc.encode(payload))

def build_incident_context():
    incident_logs = "\n".join([
        f"2026-06-23 10:14:{i:02d} payment-service ERROR checkout_id=chk_{1000+i} provider=stripe status=502 retry={i%3} message=upstream timeout from payments-api"
        for i in range(120)
    ])

    support_tickets = json.dumps([
        {
            "ticket_id": f"CS-{2000 + i}",
            "customer_tier": "enterprise" if i % 4 == 0 else "pro",
            "channel": "chat" if i % 2 == 0 else "email",
            "issue_type": "payment_failed",
            "status": "waiting_for_retry",
            "country": ["DE", "AT", "CH", "NL"][i % 4],
            "message": "I tried to pay and got an error. Please check my invoice.",
        }
        for i in range(180)
    ], indent=2)

    root_cause_notes = "\n".join([
        "Incident summary: Stripe returned intermittent 502s on the payment authorization endpoint.",
        "Impact: checkout success rate dropped from 98.7% to 81.4% for 17 minutes.",
        "Timeline:",
        "- 10:12 UTC: first spike in failed payment attempts",
        "- 10:14 UTC: customer support started receiving duplicate complaints",
        "- 10:15 UTC: retries increased load on the upstream payments API",
        "- 10:19 UTC: fallback route engaged, incident stabilized",
        "Hypothesis: upstream network issue caused transient failures, not a code regression.",
        "Next step: correlate with provider status page and retry policy metrics.",
        "Open questions: was the failure regional, and did idempotency affect retries?",
    ])

    return [
        {"role": "system", "content": "You are helping investigate a payment incident. Keep only high-signal facts."},
        {"role": "user", "content": "Please analyze the incident material and summarize what matters for the on-call response."},
        {"role": "assistant", "content": "I will inspect the logs, support tickets, and root-cause notes."},
        {"role": "tool", "tool_call_id": "logs", "content": incident_logs},
        {"role": "tool", "tool_call_id": "tickets", "content": support_tickets},
        {"role": "tool", "tool_call_id": "notes", "content": root_cause_notes},
    ]

def print_context(title, messages):
    print(title)
    print(f"TOKENS: {token_count(messages)}")
    print("CONTEXT")
    for msg in messages:
        role = msg.get("role", "unknown")
        content = str(msg.get("content", "")).replace("\n", " ")
        preview = content[:160]
        print(f"- {role}: {preview}")
```

## 2) Before Headroom

This is the raw context — lots of text, lots of repetition, lots of noise going straight into the prompt.


```python
messages = build_incident_context()
print_context("RAW", messages)
```

## 3) Apply Headroom

Now we compress the same context. For this demo we use an aggressive setting so the effect is clearly visible.


```python
messages = build_incident_context()
raw_tokens = token_count(messages)

result = compress(
    messages,
    model=MODEL,
    compress_user_messages=True,
    protect_recent=0,
    target_ratio=0.3,
    min_tokens_to_compress=0,
)

compressed_tokens = token_count(result.messages)
tokens_saved = raw_tokens - compressed_tokens
saved_pct = (tokens_saved / raw_tokens * 100) if raw_tokens else 0

print_context("COMPRESSED", result.messages)
print(f"SAVED TOKENS: {tokens_saved}")
print(f"SAVED PCT: {saved_pct:.1f}%")
print("TRANSFORMS")
print(", ".join(result.transforms_applied) if result.transforms_applied else "none")
```

## 4) Live OpenAI: Before vs. After

We send **the same prompt twice** to OpenAI — once with the raw context, once with the Headroom-compressed context.
Both replies come from OpenAI. The numbers (`prompt_tokens`, `latency`) come from the API response, not from tiktoken.
This shows: same answer quality, significantly fewer tokens, less latency.


```python
import os
import time
from dotenv import load_dotenv
from openai import OpenAI
from headroom import compress

# Rebuild context and result if this cell is run before the cells above
if "build_incident_context" not in dir() or "result" not in dir():
    import json as _json, tiktoken as _tiktoken
    _MODEL = "gpt-4o"

    def build_incident_context():
        incident_logs = "\n".join([
            f"2026-06-23 10:14:{i:02d} payment-service ERROR checkout_id=chk_{1000+i} provider=stripe status=502 retry={i%3} message=upstream timeout from payments-api"
            for i in range(120)
        ])
        support_tickets = _json.dumps([
            {"ticket_id": f"CS-{2000+i}", "customer_tier": "enterprise" if i%4==0 else "pro",
             "channel": "chat" if i%2==0 else "email", "issue_type": "payment_failed",
             "status": "waiting_for_retry", "country": ["DE","AT","CH","NL"][i%4],
             "message": "I tried to pay and got an error. Please check my invoice."}
            for i in range(180)
        ], indent=2)
        root_cause_notes = "\n".join([
            "Incident summary: Stripe returned intermittent 502s on the payment authorization endpoint.",
            "Impact: checkout success rate dropped from 98.7% to 81.4% for 17 minutes.",
            "Timeline:",
            "- 10:12 UTC: first spike in failed payment attempts",
            "- 10:14 UTC: customer support started receiving duplicate complaints",
            "- 10:15 UTC: retries increased load on the upstream payments API",
            "- 10:19 UTC: fallback route engaged, incident stabilized",
            "Hypothesis: upstream network issue caused transient failures, not a code regression.",
            "Next step: correlate with provider status page and retry policy metrics.",
            "Open questions: was the failure regional, and did idempotency affect retries?",
        ])
        return [
            {"role": "system", "content": "You are helping investigate a payment incident. Keep only high-signal facts."},
            {"role": "user", "content": "Please analyze the incident material and summarize what matters for the on-call response."},
            {"role": "assistant", "content": "I will inspect the logs, support tickets, and root-cause notes."},
            {"role": "tool", "tool_call_id": "logs", "content": incident_logs},
            {"role": "tool", "tool_call_id": "tickets", "content": support_tickets},
            {"role": "tool", "tool_call_id": "notes", "content": root_cause_notes},
        ]

    _msgs = build_incident_context()
    result = compress(
        _msgs, model=_MODEL,
        compress_user_messages=True, protect_recent=0,
        target_ratio=0.3, min_tokens_to_compress=0,
    )

load_dotenv(dotenv_path=".env", override=False)
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise RuntimeError("Missing OPENAI_API_KEY in .env")

model = os.getenv("OPENAI_MODEL", "gpt-4o-mini")
client = OpenAI(api_key=api_key)

SYSTEM = "You are a concise on-call assistant. Answer only from the provided context."
QUESTION = "Give me a 2-bullet summary: what failed, and what was the customer impact?"


def flatten_context(messages):
    """Merge all non-system messages into one text block for a fair comparison."""
    parts = []
    for msg in messages:
        role = msg.get("role", "unknown")
        content = str(msg.get("content", ""))
        if role != "system":
            parts.append(f"[{role.upper()}]\n{content}")
    return "\n\n".join(parts)


def call_openai(label, context_text):
    msgs = [
        {"role": "system", "content": SYSTEM},
        {"role": "user", "content": f"{context_text}\n\n{QUESTION}"},
    ]
    t0 = time.perf_counter()
    response = client.chat.completions.create(
        model=model,
        messages=msgs,
        max_tokens=150,
    )
    latency_ms = (time.perf_counter() - t0) * 1000
    u = response.usage
    return {
        "label": label,
        "latency_ms": round(latency_ms),
        "prompt_tokens": u.prompt_tokens,
        "completion_tokens": u.completion_tokens,
        "total_tokens": u.total_tokens,
        "reply": response.choices[0].message.content,
    }


raw_text = flatten_context(build_incident_context())
compressed_text = flatten_context(result.messages)

print(f"Calling OpenAI [{model}] with RAW context  ({result.tokens_before:,} tokens)...")
before = call_openai("RAW", raw_text)

print(f"Calling OpenAI [{model}] with HEADROOM context ({result.tokens_after:,} tokens)...")
after = call_openai("HEADROOM", compressed_text)

# --- comparison table ---
pt_saved = before["prompt_tokens"] - after["prompt_tokens"]
pt_pct = pt_saved / before["prompt_tokens"] * 100
lat_saved = before["latency_ms"] - after["latency_ms"]

print()
print(f"{'Metric':<22} {'RAW':>10} {'HEADROOM':>10} {'Saved':>10} {'%':>7}")
print("-" * 63)
print(f"{'prompt_tokens':<22} {before['prompt_tokens']:>10,} {after['prompt_tokens']:>10,} {pt_saved:>10,} {pt_pct:>6.1f}%")
print(f"{'completion_tokens':<22} {before['completion_tokens']:>10,} {after['completion_tokens']:>10,}")
print(f"{'total_tokens':<22} {before['total_tokens']:>10,} {after['total_tokens']:>10,}")
print(f"{'latency_ms':<22} {before['latency_ms']:>10,} {after['latency_ms']:>10,} {lat_saved:>10,}")

print()
print("=== RAW reply ===")
print(before["reply"])
print()
print("=== HEADROOM reply ===")
print(after["reply"])
```

## 6) Key takeaway

Headroom is not the model.
Headroom sits in front of it and turns a large, expensive context into a smaller, cheaper, more focused one.


## 7) How Headroom works internally: The three stages

**Stage 1: Cache Aligner**

Dynamic content like timestamps, UUIDs, and session tokens is extracted from the system prompt and moved to the end. This keeps the beginning of the prompt stable so caching systems (Anthropic, OpenAI) can reuse it across repeated calls.

---

**Stage 2: Smart Crusher**

The compression engine for large JSON tool outputs. It reduces large arrays to a small, representative subset — fewer tokens, key information preserved.

What it prefers to keep: first and last entries, errors, outliers, query-relevant hits, points where the pattern changes significantly.

What it removes: repeated entries, constant/boring data, bulk with no new informational value.

How it works: detects JSON arrays → analyzes content statistically → scores each item → selects the most important → compresses the rest away.

---

**Stage 3: Context Manager**

Ensures everything fits in the context window. When the conversation history grows too long, it decides what stays and what gets dropped — important context is preserved, dead weight is removed.

Two modes:
- **Rolling Window**: removes oldest messages first; system prompt and recent turns are protected; tool calls and their responses are treated as a unit.
- **Intelligent Context**: scores messages by recency, relevance, errors, references in later replies, and token density; removed content can be stored externally.

---

**In short:** Smart Crusher compresses large tool data. Context Manager keeps the whole conversation small enough for the model.

## 8) Output Token Reduction: cut what the model writes back

Every stage so far shrinks what you **send**. But you also pay for every token the model **writes back** — on Opus-class models output costs **5× more** than input.

A lot of that output is waste:
- Preambles: "Great, let me take a look at this..."
- Re-printing code you just sent it
- Deep reasoning on routine turns like reading a file or a passing test

**Verbosity steering** fixes this: a short "be terse" note is appended to the *end* of the system prompt (end = cache prefix stays intact). The model replies shorter without losing quality.

In Headroom this runs automatically via the proxy (`HEADROOM_OUTPUT_SHAPER=1`). Here we demonstrate the mechanism directly — no proxy, manually simulated.

```python
import os, time
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(dotenv_path=".env", override=False)
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
model = os.getenv("OPENAI_MODEL", "gpt-4o-mini")

BASE_SYSTEM = "You are a concise on-call assistant. Answer only from the provided context."

VERBOSITY_NOTE = (
    "\n\nBe direct and concise. Do not restate the question, repeat context already "
    "provided, add preambles like 'Great, let me...' or 'Certainly!', or include "
    "closing summaries unless explicitly asked."
)

CONTEXT = """
Incident summary: Stripe returned intermittent 502s on the payment authorization endpoint.
Impact: checkout success rate dropped from 98.7% to 81.4% for 17 minutes.
Timeline:
- 10:12 UTC: first spike in failed payment attempts
- 10:14 UTC: customer support started receiving duplicate complaints
- 10:19 UTC: fallback route engaged, incident stabilized
Hypothesis: upstream network issue, not a code regression.
"""

QUESTION = "Give me a 2-bullet summary: what failed, and what was the customer impact?"


def call(label, system_prompt):
    msgs = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": f"{CONTEXT}\n\n{QUESTION}"},
    ]
    t0 = time.perf_counter()
    resp = client.chat.completions.create(model=model, messages=msgs, max_tokens=300)
    latency_ms = round((time.perf_counter() - t0) * 1000)
    u = resp.usage
    return {
        "label": label,
        "completion_tokens": u.completion_tokens,
        "prompt_tokens": u.prompt_tokens,
        "latency_ms": latency_ms,
        "reply": resp.choices[0].message.content,
    }


normal = call("NORMAL", BASE_SYSTEM)
shaped = call("VERBOSITY STEERING", BASE_SYSTEM + VERBOSITY_NOTE)

saved = normal["completion_tokens"] - shaped["completion_tokens"]
saved_pct = saved / normal["completion_tokens"] * 100

print(f"{'Metric':<22} {'NORMAL':>10} {'SHAPED':>10} {'Saved':>10}")
print("-" * 54)
print(f"{'completion_tokens':<22} {normal['completion_tokens']:>10} {shaped['completion_tokens']:>10} {saved:>9}  ({saved_pct:.0f}%)")
print(f"{'prompt_tokens':<22} {normal['prompt_tokens']:>10} {shaped['prompt_tokens']:>10}")
print(f"{'latency_ms':<22} {normal['latency_ms']:>10} {shaped['latency_ms']:>10}")
print()
print("=== NORMAL reply ===")
print(normal["reply"])
print()
print("=== VERBOSITY STEERING reply ===")
print(shaped["reply"])
```
